```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from norm import norm
```

```python
l800 = np.genfromtxt('23_01_2020/FPI/offener_HeNe/data2_L=800mm.txt', skip_header=4).T[1]
# Peaksuche zur Skalierung
peak800 = []
for peak in find_peaks(l800)[0]:
    if l800[peak] > 100:
        peak800.append(peak)
peak800.remove(71)

fsr = 3e8/(2*8e-2)
delta_800 = peak800[1] - peak800[0]
scale800 = (np.asarray(range(len(l800))) - peak800[0]) * fsr / delta_800
#plt.scatter(peak800, [l800[peak] for peak in peak800])
plt.plot(scale800, norm(l800))
plt.xlabel('$\Delta ν [Hz]$')
plt.ylabel('a.u.')
plt.savefig('images/l800.png', dpi=300)
```

```python
# Bestimmung der Resonatorlänge
peak800_reso = []
for peak in find_peaks(l800)[0]:

    if l800[peak] > 59:
        if peak < 100 and peak > 71:
            peak800_reso.append(peak)

plt.scatter(peak800_reso, [ l800[peak] for peak in peak800_reso ])
plt.plot(range(len(l800)), l800)
delta_v_800 = peak800_reso[1] - peak800_reso[0]
delta_v_800 = delta_v_800 * fsr / delta_800
l_calc_800 = 3e8 / (2*delta_v_800)
delta_stat_800 = 4 * fsr/delta_800
stat_l_800 = l_calc_800 / delta_v_800 * delta_stat_800
print('nu_800 = {0} +- {1} ({2})'.format(delta_v_800/1e6, delta_stat_800/1e6, delta_stat_800*100/delta_v_800))
```
