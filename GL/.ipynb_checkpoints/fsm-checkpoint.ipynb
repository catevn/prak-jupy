```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

spektrum = pd.read_csv('23_01_2020/Faserspektrometer/data1.txt', encoding = "ISO-8859-1", delimiter='\t')
spektrum.plot('wavelength', 'int')
spliced = spektrum[spektrum['wavelength']<640][spektrum['wavelength']>625]
```

```python
plt.plot(spektrum['wavelength'], spektrum['int'])
plt.xlabel('Wellenlänge [nm]')
plt.ylabel('Anzahl [-]')
plt.savefig('images/fsmtotal.png', dpi=300)
```

```python
plt.plot(spliced['wavelength'], spliced['int'])
plt.xlabel('Wellenlänge [nm]')
plt.ylabel('Anzahl [-]')
plt.savefig('images/fsmspliced.png', dpi=300)
```
