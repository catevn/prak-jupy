```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# Datenimport
test = pd.read_csv('23_01_2020/Kaustik/77mm.csv', delimiter=';', encoding = "ISO-8859-1")
test.iloc[4][50]
test.iloc[4][65]

width = pd.DataFrame()
x = []
hor_width = []
ver_width = []

for file in os.listdir('23_01_2020/Kaustik/'):
    if file.endswith('.csv'):
        file_path = os.path.join('23_01_2020/Kaustik/', file)
        data = pd.read_csv(file_path, delimiter=';', encoding = "ISO-8859-1")
        x.append(float(file[:-6]))
        hor_width.append(data.iloc[4][50])
        ver_width.append(data.iloc[4][65])
```

```python
# Theoretische Berechnung
wavelength = 632.8e-6 #mm
L = 800 #mm
R = 1000 #mm
w_0 = ((wavelength/np.pi)**2 * L*(R-L))**0.25
x_0 = np.pi/(wavelength*w_0**2)

def w_prime(x, f):
    w_x = w_0 * np.sqrt(1+(x/x_0)**2)
    b = wavelength / (np.pi*w_x**2)
    b_prime =  b*((1-x/f)*(1-s/f) - (-1/f)*(s+x-s*x/f)) / ( (1-x/f)**2 + b**2 * (s+x-s*x/f)**2)
    return np.sqrt(wavelength/(np.pi*b_prime))
```

```python
s = 95 + 245 + 285 #mm
delta_x = 10*np.sqrt(4) #mm
d = np.asarray(x) + s
hor_width = 5.6e-3*np.asarray(hor_width)
ver_width = 5.6e-3*np.asarray(ver_width)
```

```python

plt.errorbar(d, hor_width, fmt='x', xerr=delta_x, label='Horizontal')
plt.errorbar(d, ver_width, fmt='x', xerr=delta_x, label='Vertikal')
plt.ylabel('w [µm]')

from scipy.optimize import curve_fit
fit, std = curve_fit(w_prime, d-s, hor_width, p0=[150])

def w_vor(s):
    return np.sqrt(wavelength*(1+s**2*b**2)/(np.pi*b))

plt.plot(np.linspace(s, 410+s, 1000), [ w_prime(i, fit[0]) for i in np.linspace(0, 410, 1000)], label='f = {0}mm (Fit)'.format(round(fit[0], 1)))
plt.plot(np.linspace(s, 410+s, 1000), [ w_prime(i, 150) for i in np.linspace(0, 410, 1000)], label='f = {0}mm'.format(150))
plt.xlabel('Weglänge Laser Kamera [mm]')
plt.legend()
plt.savefig('images/kaustik.png', dpi=300)
```
