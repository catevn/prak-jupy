```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from norm import norm
```

```python
comm =  np.genfromtxt('23_01_2020/FPI/kommerzieller_HeNe/data2_edit.txt', skip_header=1)
# Peakbestimmung
peaks = find_peaks(comm)
low_peaks = []
# Auswahl physikalisch vernÃ¼nftiger Peaks
for peak in peaks[0]:
    if comm[peak] > 50:
        low_peaks.append(peak)
low_peaks.remove(124)
```

```python
# Berechnung freier Spektralbereich
fsr = 3e8/(2*8e-2)
delta_fsr = 3e8/(2*8e-2)**2 * 0.5e-2
print('FSR = {0} +- {1}, ({2})'.format(fsr*1e-9, delta_fsr*1e-9, 100*delta_fsr/fsr))
```

```python
fsr_kanal = low_peaks[2]-low_peaks[0]
# X-Skala
scale = (np.asarray(range(len(comm)))-low_peaks[0]) * fsr/fsr_kanal
plt.plot(scale, norm(comm))
plt.scatter([scale[peak] for peak in low_peaks], [ norm(comm)[peak] for peak in low_peaks], c='red', marker='x', label='Peaks')
plt.xlabel('$\Delta Î½$ [Hz]')
plt.vlines(scale[low_peaks[0]], 0, 1, alpha=0.4, color='g', label='FSR')
plt.vlines(scale[low_peaks[2]], 0, 1, alpha=0.4, color='g')
# Berechnung der Halbwertsbreite
halb = comm[low_peaks[1]]/2
np.argmin(abs(np.asarray(comm)-halb))
plt.scatter(scale[231], norm(comm)[231], label='FWHM', c='y')
plt.scatter(scale[216], norm(comm)[216], c='y')
plt.legend()
plt.ylabel('a.u.')
plt.savefig('images/kommerz_komplett.png', dpi=300)
fwhm = 2*(scale[low_peaks[1]]-scale[216])
# Abstand zweier Maxima
delta = scale[low_peaks[1]] - scale[low_peaks[0]]
finesse = fsr/fwhm
# Messunsicherheit Finesse
delta_finesse = 4*fsr/fsr_kanal / fwhm
print('F = {0} +- {1} ({2}%)'.format(finesse, delta_finesse, 100*delta_finesse/finesse))
```

```python
# Berechnung der ReflektivitÃ¤t
a = finesse/np.pi
xi = (np.sqrt(1+4*a**2) - 1)/(2*a)
r = xi**2
r
dr_dfinesse = 2*xi*(2/(np.sqrt(1+4*a**2)) - (np.sqrt(1+4*a**2) -1)/(2*a**2))/np.pi
delta_r = np.sqrt(dr_dfinesse**2 * delta_finesse**2)
print('R = {0} +- {1} ({2}%)'.format(r, delta_r, 100*delta_finesse/finesse))
```
