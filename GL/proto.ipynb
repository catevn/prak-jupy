```python
# 1. Bestimmung der Stabilitätsparameter:
# Planparalleler Spiegel:
g_p = 1
g_hohl = 1 - 500/1000
```

```python
# 2. Justage & Verstärkung
# Zur Bestimmung der Verstärkung wird ein komerzieller He-Ne-Laser verwendet, welcher durch Justage in die Gasröhre geführt wird.
# Weiterhin wird der grüne Laser in entgegengesetzer Achsenrichtung justiert.

# Kommerziell / Inaktiv / Aktiv / Hintergrund
# N Anzahl Messungen
N_2 = [61, 61, 61, 61]
# Leistung
P_2 = [1.3, 0.936, 1.1, 1.826e-3] #mW
# Standardabweichung
P_std_2 = [3.9, 10.3, 10.1, 36e-3] # µW
P_2_syst = [0.05, 0.0005, 0.05, .5e-6] #mW
labels = ['Ohne Röhre', 'Röhre inaktiv', 'Röhre aktiv']
delta_p_2_stat = 1e-3*np.asarray(P_std_2)/np.sqrt(61)

plt.bar(labels, np.asarray(P_2[:-1])-P_2[3], width=0.5, yerr = np.sqrt(np.asarray(delta_p_2_stat[:-1])**2 + delta_p_2_stat[3]**2 + np.asarray(P_2_syst[:-1])**2 + P_2_syst[3]**2))
plt.ylabel('$P - P_{Hintergrund}$ [mW]')
plt.ylim(0.8, 1.4)
plt.savefig('images/verstaerkerwerte.png', dpi=300)
```

```python
# Berechnung der Verstärkung
v = P_2[2]/P_2[1]
delta_v_stat = np.sqrt(delta_p_2_stat[2]**2/P_2[1] +  delta_p_2_stat[1]**2*(P_2[2]/P_2[1]**2)**2)
delta_v_syst = np.sqrt(P_2_syst[2]**2/P_2[1] +  P_2_syst[1]**2*(P_2[2]/P_2[1]**2)**2)
rel_error = np.sqrt(delta_v_stat**2+delta_v_syst**2)/v
print('Verstärkung V = {0}'.format(v))
print('stat. Messunsicherheit Verstärkung = {0}'.format(delta_v_stat))
print('syst. Messunsicherheit Verstärkung = {0}'.format(delta_v_syst))
print('Relativer Messfehler = {0}'.format(rel_error))
```

```python
# 3. Bauen des Resonators
# Zuerst werden die beiden Resonatorspiegel, ein Planparalleler und ein Hohlspiegel, als hemishärischer Resonator verbaut.
# Diese werden so justiert, dass die Rückreflexe auf die Reflexe der Justagelaser fallen.
# Durch 'Beamwalken' wird die Leistung optimiert.
```

```python
# Leistung in Abhängigkeit der Resonatorlänge
# Nullmessung P_0
P_0 = P_2[3]
P_0_err = delta_p_2_stat[3]
L = [60, 90, 95, 100, 100.5, 80] #cm
N_3a = [98, 61, 62, 60, 61, 79]
P_max = [0.970, 0.865, 0.725, 0.4, 290e-3, 1.02] #mW
P_max_std = [4.5, 3.7, 3.8, 1.6, 8.6, 2.6] #µW
P_max_err = 1e-3*np.asarray(P_max_std)/np.sqrt(N_3a)
plt.errorbar(L, np.asarray(P_max)-P_0, fmt='x', xerr=0.5, yerr=np.sqrt(np.asarray(P_max_err)**2 + P_0_err**2))
plt.xlabel('L [cm]')
plt.ylabel('$P - P_{Hintergrund}$ [mW]')
plt.savefig('images/laenge.png', dpi=300)
```

```python
# Malus law, L=800mm
from scipy.optimize import curve_fit
def cos_fit(x, a, b, c):
    return a*np.cos(b*x*np.pi/180 + c)**2
theta = [271, 286, 301, 316, 331, 346, 1, 31, 61, 91]
# Reskalieren der Winkel für den Plot
for i in range(6):
    theta[i] = theta[i] - 180

N_3b = [34, 31, 31, 30, 30, 31, 30, 31, 30, 30]
P_malus = [3.04e-3, 51.5e-3, 192e-3, 0.386, 0.585, 0.727, 0.779, 0.58, 194e-3, 3.31e-3] #mW
P_malus_std = [177e-3, 0.179, 0.606, 2.3, 2.1, 2.9, 1.8, 1, 0.9, 13e-3] #µW

plt.errorbar(np.asarray(theta), np.asarray(P_malus)-P_0, fmt='x', yerr=1e-3*np.asarray(P_malus_std)/np.sqrt(N_3b), xerr=2)
plt.xlabel('Winkel [°]')
plt.ylabel('$P - P_{Hintergrund}$ [mW]')
fit, std = curve_fit(cos_fit, theta, P_malus, p0=[0.8, 1, 0])
plt.plot(np.linspace(0, 180, 100), [ cos_fit(i, fit[0], fit[1], fit[2]) for i in np.linspace(0, 180, 100)])
plt.savefig('images/winkel.png', dpi=300)
```
