```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.optimize import curve_fit
from norm import norm
```

```python
l800 = np.genfromtxt('23_01_2020/FPI/offener_HeNe/data2_L=800mm.txt', skip_header=4).T[1]
# Peaksuche zur Skalierung
peak800 = []
for peak in find_peaks(l800)[0]:
    if l800[peak] > 100:
        peak800.append(peak)
peak800.remove(71)

fsr = 3e8/(2*8e-2)
delta_800 = peak800[1] - peak800[0]
scale800 = (np.asarray(range(len(l800))) - peak800[0]) * fsr / delta_800
plt.scatter((np.asarray(peak800) - peak800[0])*fsr/delta_800, [l800[peak] for peak in peak800])
plt.plot(scale800, l800)
plt.xlim(-0.5e9, 0.5e9)
plt.xlabel('$\Delta Î½ [Hz]$')
```

```python
peaks_for_fit = []
for peak in find_peaks(l800)[0]:
    if peak < 150:
        if l800[peak] > 61:
            peaks_for_fit.append(peak)
plt.plot(np.asarray(range(len(l800))), l800)
plt.scatter(np.asarray(peaks_for_fit), [l800[peak] for peak in peaks_for_fit])
```

```python
def gauss(x, mu, sigma, a):
    return a*np.exp(-(x-mu)**2/(2*sigma**2))
fit, std = curve_fit(gauss, peaks_for_fit, [l800[peak]-min(l800) for peak in peaks_for_fit], p0=[70, 50, 100])
plt.plot((np.asarray(range(len(l800)))-peaks_for_fit[1])*fsr/delta_800, norm(l800-min(l800)), label='Messwerte')
plt.plot((np.linspace(0, 200, 1000) - peaks_for_fit[1] )*fsr/delta_800, np.asarray([ gauss(i, fit[0], fit[1], fit[2]) for i in np.linspace(0, 200, 1000)])/(l800-min(l800)).max(), label='Fit')
plt.xlabel('$\Delta Î½$ [Hz]')
plt.xlim(-.5e9, .5e9)
plt.ylabel('a.u.')
sigma_hz = fit[1] * fsr/delta_800
plt.legend()
plt.savefig('images/enclosed.png', dpi=300)
sigma_hz/1e6
delta_sigma = np.sqrt(std[1][1])*fsr / delta_800 / 1e6
fwhm = 2*np.sqrt(2*np.log(2)) * sigma_hz
fwhm_delta = delta_sigma * 2*np.sqrt(2*np.log(2))
fwhm/1e6
fwhm_delta/1e6
```

```python
# Temperatur
from scipy.constants import c
from scipy.constants import k
from scipy.constants import u

nu_0 = c/632.8e-9
m = 20.1797*u
T = (fwhm)**2 * m * c**2 /(4*nu_0**2 * 2 * np.log(2)*k)
delta_T = np.sqrt((2*T/fwhm * fwhm_delta)**2)
print('Temperatur = {0} +- {1}, {2}'.format(T, delta_T, delta_T/T*100))
```
